## Looking at BMI vs. Survival

```python
cols = ['age', 'sex', 'bmi', 'asa', 'department', 'approach']

for col in cols:
    plt.hist(data[col], bins=30)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title(f"Distribution of Patient {col}")
    plt.show()
```

```python
plt.scatter(data['age'], data['bmi'], alpha=0.4)
plt.xlabel("Age")
plt.ylabel("BMI")
plt.title("Age vs BMI")
plt.show()
```

```python
plt.scatter(data['age'], data['icu_days'], alpha=0.4)
plt.xlabel("Age")
plt.ylabel("ICU Days")
plt.title("Age vs ICU Days")
plt.show()
```

```python
mortality_counts = data['death_inhosp'].value_counts().sort_index()
plt.bar(['Survived', 'Died'], mortality_counts.values)
plt.ylabel("Number of Patients")
plt.title("In-Hospital Mortality Count")
plt.show()
```
